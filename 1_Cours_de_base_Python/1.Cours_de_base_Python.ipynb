# 4. Les fonctions 
En Python, une fonction est un bloc de code réutilisable.
On lui donne un nom, ce qui permet de l’appeler facilement plus tard.
On peut aussi lui passer des paramètres, c’est-à-dire des informations que la fonction utilise pour effectuer ses calculs ou afficher des résultats.

L’avantage principal des fonctions est qu’on peut les exécuter autant de fois qu’on veut, sans avoir à réécrire le même code à chaque fois.

Pourquoi utiliser des fonctions ?
- Elles permettent d’éviter de répéter plusieurs fois le même code.
- Elles rendent le programme plus clair, plus organisé et plus facile à modifier ou à étendre.

En pratique, grâce aux fonctions, votre code devient plus propre, lisible et modulable, ce qui est essentiel dès que vos programmes deviennent un peu plus longs ou complexes.

```python
# Exemple: Définir la fonction
def eau_restante(pluie, evaporation):
    """
    Calcule l'eau restante après la pluie et l'évaporation.
    """
    return pluie - evaporation

# Utiliser la fonction
reste1 = eau_restante(50, 20)
reste2 = eau_restante(30, 15)
reste3 = eau_restante(10, 5)

print("Eau restante 1 :", reste1, "mm")
print("Eau restante 2 :", reste2, "mm")
print("Eau restante 3 :", reste3, "mm")
```

# 5. Conditions: if
Un if / else est une structure conditionnelle.
Elle permet à ton programme de prendre une décision :
- si (if) une condition est vraie → on exécute un bloc de code
- sinon (else) → on exécute un autre bloc de code

```python
# Exemple: 
pluie = 8  # en millimètres
if pluie >= 10:
    print("Il pleut beaucoup.")
else:
    print("Il pleut peu.")
```

```python
# Exemple avec plusieurs conditions (elif)
pluie = 3

if pluie < 5:
    print("Presque pas de pluie.")
elif pluie < 20:
    print("Il pleut un peu.")
else:
    print("Très forte pluie !")
```

# 6. L’utilisation d’une boucle (loop)

Une boucle est une structure qui permet de répéter automatiquement une série d’instructions plusieurs fois, sans avoir à les réécrire.

Comment ça fonctionne :
- On définit la liste de valeurs sur laquelle on veut itérer.
- Pour chaque élément de cette liste, le code à l’intérieur de la boucle est exécuté.
- On peut utiliser la valeur de l’élément à l’intérieur de la boucle pour effectuer des calculs ou afficher des informations.

```python
# Exemple d'une boucle
# Liste de pluie en mm pour 3 jours
pluie = [5, 10, 3]

# Boucle pour afficher la pluie chaque jour
for valeur in pluie:
    print("Pluie :", valeur, "mm")
```

```python
# Exemple d'une boucle
# Liste de pluie en mm sur 5 jours
pluie_jours = [5, 12, 0, 8, 15]

# Initialiser le total
total_pluie = 0

# Boucle pour parcourir chaque jour
for i, valeur in enumerate(pluie_jours, start=1):							# permet d’avoir la valeur et le numéro du jour en même temps.
    total_pluie += valeur  													# additionne progressivement la pluie pour obtenir un total courant
    print(f"Jour {i} : {valeur} mm, Total jusqu'ici : {total_pluie} mm")

# Afficher le total final
print(f"Total de pluie sur 5 jours : {total_pluie} mm")
```
