# 4. Functions
In Python, a function is a reusable block of code. A function gets a name, which allows us to call it easily later in the code.
We also enter parameters into a function. These parameters are pieces of information that a function uses to perform calculations or display results. The main advantage of functions is that we can a certain process many times without rewriting the same code each time.

Why use functions?
- They help avoid repeating the same code multiple times.
- They make the program clearer, more organized, and easier to modify or extend.

In practice, thanks to functions, your code becomes cleaner, more readable, and modular, which is essential once your programs become a bit longer or more complex.

```python
# Example: Defining a function
def net_precipitation(rain, evaporation):
    """
    Calculates the remaining water after rain and evaporation.
    """
    return rain - evaporation

# Using the function
remaining1 = net_precipitation(50, 20)
remaining2 = net_precipitation(30, 15)
remaining3 = net_precipitation(10, 5)

print("Net precipitation event 1:", remaining1, "mm")
print("Net precipitation event 2:", remaining2, "mm")
print("Net precipitation event 3:", remaining3, "mm")
```

# 5. Conditions: if
An if / else is a conditional structure.
It allows your program to make a decision:
- if a condition is true → execute a block of code
- else → execute another block of code

```python
# Example:
rain = 5                # in millimeters
if rain >= 10:
    print("It is raining a lot.")
else:
    print("It is raining a little.")
```

```python
# Example with multiple conditions (elif)
rain = 3

if rain < 2:
    print("Hardly any rain.")
elif rain < 8:
    print("It is raining a little.")
else:
    print("Heavy rain!")
```

# 6. Using a loop

A loop is a structure that allows you to automatically repeat a series of instructions multiple times without having to rewrite them.

# How it works:
- Define the list of values over your would like to interate.
- For each value in this list, the code inside the loop is executed.
- You can use the value of the element inside the loop to perform calculations or display information.

```python
# Example of a loop
# List of rainfall in mm for 3 days
rain = [5, 10, 3]

# Loop to display the rainfall each day
for value in rain:
    print("Rain:", value, "mm")
```

```python
# Example of a loop

# List of rainfall in mm over 5 days
rain_days = [5, 12, 0, 8, 15]

# Initialize the total, we start with 0 mm
total_rain = 0

# Loop through each day
for i, value in enumerate(rain_days, start=1):      # allows to get both the value and the day number
    total_rain += value                           # progressively adds the rainfall to get a running total
    print(f"Day {i}: {value} mm, cumulative rainfall = {total_rain} mm")

# Display the final total
print(f"Total rainfall over 5 days: {total_rain} mm")
```
