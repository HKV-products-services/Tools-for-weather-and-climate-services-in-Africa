```python
# Solution to Exercise 3.3

# Import the plotting library
import matplotlib.pyplot as plt

# Rainfall data over several days (in mm)
days = [1, 2, 3, 4, 5, 6, 7]
rain = [5, 12, 0, 8, 15, 3, 10]

# Create a bar chart
plt.bar(days, rain, color='skyblue')

# Add a title and labels for the axes
plt.title("Rainfall over 7 days")
plt.xlabel("Day")
plt.ylabel("Rain (mm)")

# Display the plot
plt.show()
```

```python
# Solution Exercise 4.1
def total_rain(rain_list):
    """
    Calculates the total rainfall over several days.
    """
    total = sum(rain_list)
    return total

rain_days1 = [5, 12, 0, 8, 15]
rain_days2 = [3, 7, 2, 0, 10]

total1 = total_rain(rain_days1)
total2 = total_rain(rain_days2)

print(f"Total rain day 1: {total1} mm")
print(f"Total rain day 2: {total2} mm")
```

```python
# Solution Exercise 5.1

# River flows (in m³/s)
river_flow_1 = 100
river_flow_2 = 50

# Calculate the total flow
total_flow = river_flow_1 + river_flow_2
print(f"Total flow: {total_flow} m³/s")

# Check the flow with if / elif / else
if total_flow < 100:
    print("It is dry")
elif total_flow <= 200:  # between 100 and 200 inclusive
    print("Normal flow")
else:
    print("Lots of water!")
```

```python
# Solution Exercise 6.2
evaporation = [2, 5, 1, 3, 4, 2, 3]

total_evaporation = 0
number_of_days = len(evaporation)

# Loop through each day and accumulate the total
for i, value in enumerate(evaporation, start=1):
    print(f"Day {i}: {value} mm of evaporation")
    total_evaporation += value

# Calculate the average
average = total_evaporation / number_of_days

# Display the results
print(f"Total evaporation over {number_of_days} days: {total_evaporation} mm")
print(f"Average daily evaporation: {average:.2f} mm")
```
